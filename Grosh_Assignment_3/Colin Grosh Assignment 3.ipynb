## 1. Load in the Data. The Response Variable is the Churn variable and the Predictor variables are Gender, Income, Age, FamilySize, Education, Calls, and Visits variables 

```python
# Import the Necessary Libraries 
from knn import *
import numpy as np
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from util import cat_features
from sklearn.metrics import mean_squared_error, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.metrics import jaccard_similarity_score
from scipy.spatial import distance 
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import SelectPercentile 
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import RFECV

# Load in the data 
data = pd.read_csv('assignment3_data/churn_data.csv')

lst_x = list(data.columns)

# Remove the Reponse variable and ID variable 
lst_x.remove('Churn')
lst_x.remove('CustID')
#lst_x.remove('Age')

# Seperate into Explanatory and Reponse variables 
data_x = data[lst_x]
data_y = data['Churn']
```

### Split the Data into Train Test sets and Create the Preprocessing Pipeline

```python
# Create the Label Encoding Pipeline 
le = preprocessing.LabelEncoder()

# Pass the Categorical X values through the pipeline 
data_x['Gender'] = le.fit_transform(data_x['Gender'])
data_x['Income'] = le.fit_transform(data_x['Income'])

# Pass the Response variable through the pipeline 
data_y = le.fit_transform(data_y)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data_x, data_y, test_size=0.3, random_state=4)
```

## Fit different K values ad different Distance Functions on a base model and get results 

## 4. The error metrics being used are Accuracy Score, Precision Score, Recall Score and the F1 Score

```python
# Function to Determine Euclidean Distance between 2 arrays 
def euclidean(x1, x2):
    dist = math.sqrt(sum([(a-b)**2 for a, b in zip(x1, x2)]))
    return dist

# List of K values to test with  
ks = [1, 3, 5, 7, 9, 11, 13, 15]

euc_acc = []
euc_prec = []
euc_rec = []
euc_f1 = []
euc_roc_auc = []

# Loop through and fit and check the model on each k 
print("EUCLIDEAN DISTANCE")
for k in ks:
    # Create the instance of the model 
    mod = KNN(k, euclidean)
    
    # Fit the model 
    mod.fit(x_train, y_train)
    
    # Make predictions on the testing data 
    y_hat = mod.predict(x_test)
    
    # Check the different error metrics for the predictions 
    print('--------Evaluating ModeK: k = ' + str(k) + ' with the euclidean distance-------------')
    print('Accuracy: ' + str(accuracy_score(y_test, y_hat)))
    euc_acc.append(accuracy_score(y_test, y_hat))
    print('Precision: ' + str(precision_score(y_test, y_hat)))
    euc_prec.append(precision_score(y_test, y_hat))
    print('Recall: ' + str(recall_score(y_test, y_hat)))
    euc_rec.append(recall_score(y_test, y_hat))
    print('F1: ' + str(f1_score(y_test, y_hat)))
    euc_f1.append(f1_score(y_test, y_hat))
   


    
print('\n')
print("JACCARD DISTANCE")
for k in ks: 
    
    # Create the instance of the model using the Jacard Distance 
    mod = KNN(k, jaccard_similarity_score)
    
    # Fit the model 
    mod.fit(x_train, y_train)
    
    # Make predictions on the testing data 
    y_hat = mod.predict(x_test)
    
    # Check the different error metrics for the predictions 
    print('--------Evaluating ModeK: k = ' + str(k) + ' with the jaccard distance-------------')
    print('Accuracy: ' + str(accuracy_score(y_test, y_hat)))
    print('Precision: ' + str(precision_score(y_test, y_hat)))
    print('Recall: ' + str(recall_score(y_test, y_hat)))
    print('F1: ' + str(f1_score(y_test, y_hat)))
                       
print('\n')
print("COSINE DISTANCE")
for k in ks: 
    
    # Create the instance of the model using the Jacard Distance 
    mod = KNN(k, distance.cosine)
    
    # Fit the model 
    mod.fit(x_train, y_train)
    
    # Make predictions on the testing data 
    y_hat = mod.predict(x_test)
    
    # Check the different error metrics for the predictions 
    print('--------Evaluating ModeK: k = ' + str(k) + ' with the Cosine-------------')
    print('Accuracy: ' + str(accuracy_score(y_test, y_hat)))
    print('Precision: ' + str(precision_score(y_test, y_hat)))
    print('Recall: ' + str(recall_score(y_test, y_hat)))
    print('F1: ' + str(f1_score(y_test, y_hat)))
   
```
